```python
# mengatur karakter ( indexing )

nama = "Python"
print(nama[0])
print(nama[-1])
```

```python
# memotong stricing ( slicing )

teks = "belajar python"
print(teks[:])
print(teks[0:7])
print(teks[8:])
print(teks[6:14])
```

```python
# daftar string methods


# 1. ubah huruf
teks = "halo python"
print(teks.upper())
print(teks.upper())
print(teks.title())
print(teks.capitalize())


# 2. hapus spasi
teks2 = "   halo dunia   "
print(teks2.strip()) 
print(teks2.lstrip())
print(teks2.rstrip())


# 3. menelesuri huruf
nama = "samuel tambunan"
jumlah_k = nama.count("a")
print(jumlah_k)


# ganti teks
teks3 = "aku sayang kamu"
print(teks3.replace("kamu", "kevin"))


# cari teks
teks4 = "belajar python"
print(teks4.find("belajar"))
```

```python
# escape characters


# baris baru ( \n )
print("Halo\nDunia")

# tab ( \t )
data = "nama :\tSamuel Tambunan\numur :\t19"
print(data)

# kutip dalam string
print("Dia berkata: \"Aku suka Python\"")
print('Hari ini aku nonton film \'Naruto\'')
```

```python
# f-string dengan expressions


# variabel biasa
nama = "kevin"
umur = 10
print(f"nama saya {nama}, umur saya {umur}")

# ekspresi matematika
x = 10
y = 5
hasil = x + y
print(f"hasil penjumlahan {x} + {y} adalah {hasil}")

# ekspresi total
harga = 10000
jumlah = 3
total = f"total harga : Rp {harga * jumlah}"
print(total)

# ekspresi memakai upper
nama = "samuel"
salam = f"halo {nama.upper()}!"
print(salam)
```
