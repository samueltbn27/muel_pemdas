```python
# percabangan -> cara program mengambil keputusan.

# ðŸ“Œ ringkas 
#     if = cek 1 kondisi.
#     if - else = ada 2 pilihan (True atau False).
#     if - elif - else = ada banyak pilihan.
#     Bisa juga ditulis singkat pakai ternary.
```

```python
# if kondisi

angka = int(input("masukkan angka : "))

if angka > 0:
  print("angka positif")
if angka < 0:
  print("angka negatif")
if angka == 0:
  print("angka nol")
```

```python
# if else

nilai = int(input("masukkan nilai : "))

if nilai >= 60:
  print("lulus")
else:
  print("tidak lulus")
```

```python
# if elif else

nilai = 80

if nilai >= 90:
    print("A")
elif nilai >= 75:
    print("B")
elif nilai >= 60:
    print("C")
else:
    print("D")
```

```python
# Operator logika


# and
umur = 19
punya_ktp = True

if umur >= 18 and punya_ktp:
  print("boleh buat sim")
else:
  print("belum boleh buat sim")
  
# or
uang = 110000
punya_kupon = False

if uang >= 100000 or punya_kupon:
  print("boleh masuk konser")
else:
  print("tidak boleh masuk konser")
  
# not
hujan = False

if not hujan:
  print("boleh main bola di luar")
else:
  print("lebih baik di rumah")
  
# gabungan
umur = 19
punya_sim = False
punya_ktp = True

if (umur >= 18 and punya_ktp) or punya_sim:
    print("Boleh mengemudi")
else:
    print("Tidak boleh mengemudi")
```

```python
# nested if 

username = "muel"
password = 1234

if username == "muel":
  if password == 1234:
    print("login berhasil")
    print("selamat datang admin")
  else:
    print("password salah")
else:
  print("username tidak ditemukan")
```

```python
# match case

hari = "senin"

match hari:
  case "senin":
    print("hari kerja, semangat !")
  case "sabtu" | "minggu":
    print("hari libur, santai dulu")
  case _:
    print("hari biasa")

# angka 
angka = 2

match angka:
    case 1:
        print("Satu")
    case 2:
        print("Dua")
    case 3:
        print("Tiga")
    case _:
        print("Angka lain")

# nilai 
nilai = 85

match nilai:
    case n if n >= 90:
        print("A")
    case n if n >= 75:
        print("B")
    case n if n >= 60:
        print("C")
    case _:
        print("D")
```

```python
# match case ( and or not )

# and
umur = 20
punya_ktp = True

match umur:
  case u if u >= 18 and punya_ktp:
    print("boleh buat SIM")
  case _:
    print("tidak boleh buat SIM")
    
    
# or
uang = 50000  
punya_kupon = True

match uang:
  case u if u >= 100000 or punya_kupon:
    print("boleh masuk konser")
  case _:
    print("tidak boleh masuk konser")


# not
hujan = False

match hujan:
  case h if not h:
    print("cuaca cerah, ayo main bola")
  case _:
    print("hujan, mending dirumah")

# gabungan
umur = 17
punya_sim = False
punya_ktp = True

match umur:
    case u if (u >= 18 and punya_ktp) or punya_sim:
        print("Boleh mengemudi")
    case _:
        print("Tidak boleh mengemudi")
```

```python
# ternary operation

# cek angka ganjil genap
angka = 8
hasil = "Genap" if angka % 2 == 0 else "Ganjil"
print(hasil)

# nested bersarang
nilai = 90
grade = "A" if nilai >= 90 else "B" if nilai >= 75 else "C"
print(grade)

```
