```python
# List adalah salah satu struktur data dalam bahasa pemrograman python yang mampu menyimpan kumpulan data (objek/nilai), yang disebut elemen list

suhu1 = 21
suhu2 = 20
suhu3 = 22

suhu_ruangan = [suhu1, suhu2, suhu3]
print(suhu_ruangan)
```

```python
# List dapat diisi dengan tipe data apa saja, string, integer, float, boolean, object, dan sebagainya.

# membuat tipe data yang sama
buah = ["jeruk", "mangga", "apel"]
bilangan_prima = [2,3,5,7,11,13,17,19]

# membuat list tipe data berbeda
biodata_ucup = ["ucup", 19 , "bandung", "sepak bola"]

# membuat data list dicampur dengan boolean
ucup = [biodata_ucup, True]
print(ucup)
```

```python
# isi list dapat dirubah (mutable) yang artinya elemen list dapat diganti nilai elemen dengan cara ditambah maupun dikurangi.

# mengakses element dari list
tinggi_badan = [162, 177, 182, 150, 166]
print(tinggi_badan[0])

# slicing (memotong) element dari list
negara = ["german", "belanda", "amerika", "swiss"]
print(negara[0:2])

number = [0, 1, 2, 3, 4, 5]
# print(number[0:]) or print(number[:])
# print(number[2:4])

# number.reverse()
# print(number)

# list[start : stop : step]
print(number[5:1:-1]) 


#data
suhu_keluarga_ucup = ['ayah ucup', 19, 'ucup', 19, 'ibu ucup', 20]
# mengubah suhu
suhu_keluarga_ucup[3] = [20] 
suhu_keluarga_ucup[1] = [20]
print(suhu_keluarga_ucup)

suhu_keluarga_ucup = suhu_keluarga_ucup + ['adik ucup', 20]
print(suhu_keluarga_ucup)
 
```

```python
# menambahkan element list

# prepend(item) -> menambahkan item dari depan
# append(item) -> menambahkan item dari belakang

murid = ['Ayu', 'Bagus', 'Fajar', 'Cahyo', 'Ani']
murid.prepend("Rizky")  
murid.append("Sanji")


print(murid)
```

```python
# insert() untuk menambahkan item pada indeks tertentu.
murid = ['Ayu', 'Bagus', 'Fajar', 'Cahyo', 'Ani']
murid.insert(2, "Sanji")
print(murid)
```
