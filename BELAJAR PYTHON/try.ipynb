```python
usia = int(input("masukkan usia anda : "))

if usia <= 12:
  harga_tiket = 30000
elif  12 <= usia <= 17:
  harga_tiket = 40000
else:
  harga_tiket = 50000

print(f"harga tiket anda : {harga_tiket}")
```

```python
kendaraan = input("kendaraan anda (mobil / motor) ?")
jam = int(input("lama parkir : "))

if kendaraan == "motor":
  biaya_parkir = jam * 2000
elif kendaraan == "mobil":
  biaya_parkir = jam * 5000
else:
  print("tidak valid")
  
print(f"biaya : {biaya_parkir}")
```

```python
total_belanja = int(input("total belanja :"))

diskon = 0
if total_belanja >= 500.000:
  diskon =  0.20
elif total_belanja >= 250000:
  diskon =  0.10
else:
  diskon = 0

harga_diskon = total_belanja * diskon
total_keseluruhan = total_belanja - harga_diskon

print(f"total biaya keseluruhan : {total_keseluruhan}")
```

```python
angka = int(input("masukkan angka anda :"))
angka_rahasia = 7

if angka == angka_rahasia:
  print("selamat, tebakan anda benar")
elif angka < angka_rahasia:
  print("angka anda terlalu kecil")
else:
  print("angka anda terlalu besar")
```

```python
angka_rahasia = 15
kesempatan = 4

for i in range(kesempatan):
  angka = int(input("masukkan angka : "))
  
  if angka == angka_rahasia:
    print("selamat tebakan benar")
  else:
    print("tebakan salah")

if angka != angka_rahasia:
  print("kesempatan telah habis coba lagi nanti")
```
