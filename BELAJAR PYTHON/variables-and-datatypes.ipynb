```python
# tipe data -> jenis isi kotak ( variabel ) tersebut.


# 1. integer ( angka bulat )
umur = 19

# 2. float ( angka desimal )
tinggi = 170.5
speed = 3e8 

# 3. string ( teks/tulisan )
nama = "kevin"
puisi = "ini adalah puisi"

# 4. boolean ( benar/salah )
mahasiswa = True
lulus = True

# bilangan besar ( python bisa handle angka sangat besar )
populasi_dunia = 78000000
angka_besaar = 12345678910
print(type(populasi_dunia))
print(type(angka_besaar))
```

```python
# membuat variabel 


# multiple assignment 
a, b, c = 10, 20, 30
x = y = z = 100

a = 5
b = 10
a, b = b, a

# mengubah nilai variabel
skor = 80
print(80)
skor = 95
print(skor)
```

```python
# menggunakan variabel


# menggunakan variabel dalam operasi
nama_depan = "Samuel"
nama_belakang = "Tambunan"
nama_lengkap = nama_depan +  " " + nama_belakang
print(nama_lengkap)

panjang = 5
lebar = 10
luas = panjang * lebar
print(luas)

# menggunakan variabel dalam print
print("nama lengkap :", nama_lengkap)
print("panjang =", panjang, "lebar =", lebar, "panjang =", luas)
```
