```python
# operator -> simbol yang dipakai untuk membuat operasi perhitungan pada nilai

# jadi singkatnya:
#   Aritmatika → hitungan angka
#   Perbandingan → membandingkan nilai
#   Logika → True/False
#   Penugasan → isi variabel
#   Keanggotaan → cek di dalam list/string
#   Identitas → cek objek sama/tidak
```

```python
# operator aritmatika

a = 10
b = 3

print(a + b)   # 13 (tambah)
print(a - b)   # 7  (kurang)
print(a * b)   # 30 (kali)
print(a / b)   # 3.333... (bagi biasa, hasil float)
print(a // b)  # 3 (bagi bulat, dibulatkan ke bawah)
print(a % b)   # 1 (sisa bagi / modulo)
print(a ** b)  # 1000 (pangkat → 10^3)  
```

```python
# operator penugasan ( assignment )

x = 10   # isi 10
x += 5   # x = x + 5 → 15
x -= 3   # x = x - 3 → 12
x *= 2   # x = x * 2 → 24
x /= 4   # x = x / 4 → 6.0
```

```python
# operator perbandingan

x = 5
y = 10
print(x == y)  # False (sama dengan)
print(x != y)  # True  (tidak sama dengan)
print(x > y)   # False (lebih besar)
print(x < y)   # True  (lebih kecil)
print(x >= y)  # False (lebih besar/sama dengan)
print(x <= y)  # True  (lebih kecil/sama dengan)

nama1 = "alice"
nama2 = "bob"
nama3 = "alice"
print(nama1 == nama2)
print(nama1 == nama3)
```

```python
# operator logika

a = True
b = False
print(a and b)  # False (keduanya harus True)
print(a or b)   # True  (salah satu True)
print(not a)    # False (kebalikan)


umur = 25
print(umur > 18 and umur < 30)

hari = "sabtu"
print(hari == "sabtu" or hari == "minggu")

aktif = True
print(not aktif)
```

```python
# operator keanggotaan

angka = [1, 2, 3, 4, 5]

print(3 in angka)      # True  (3 ada di dalam list)
print(10 not in angka) # True  (10 tidak ada di list)
```

```python
# operator identitas 

a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)   # True  (karena b = a)
print(a is c)   # False (walaupun isinya sama, objeknya beda)
```
