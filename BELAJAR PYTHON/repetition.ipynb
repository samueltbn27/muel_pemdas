```python
# for loop -> Digunakan untuk mengulang sesuatu yang sudah jelas jumlahnya.

# range angka 0 sampai 4
for i in range(5):
  print(i)

# range angka 1 sampai 5
for i in range(1, 6):
  print(i)
  
# range "hello" sebanyak 3 kali
for i in range(3):
  print("hello")

# menghitung mundur
for i in range(1, 6, +1):
  print(i)
```

```python
# for loop dengan string

nama = input("masukkan nama anda : ")
for huruf in nama:
  print("-", huruf)
```

```python
# while loop -> Digunakan untuk mengulang selama kondisi benar (True).

angka = 5
while angka >= 1:
  print(angka)
  angka -= 1

password = ""
while password != "123456":
  password = input("masukkan password")
  if password != "123456":
    print("password salah")
print("password benar")
```

```python
# while loop ( break and continue )

# break
angka_rahasia = 7

while True:
  angka = int(input("masukkan angka : "))
  if angka == angka_rahasia:
    print("selamat anda benar")
    break
  else:
    print("salah, coba lagi")
    
# continue
for i in range(10):
  if i % 2 == 0:
    continue
  print(i)
  
```
